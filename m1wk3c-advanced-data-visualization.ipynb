### 1. **Seaborn Library Overview**

**Seaborn** is a powerful Python data visualization library built on top of Matplotlib. It provides a high-level interface to create informative and attractive statistical plots with minimal code. Seaborn comes with several benefits:
- Better visual aesthetics.
- Easy management of complex visualizations (like facet grids, pair plots, etc.).
- Integration with Pandas, allowing direct plotting from DataFrames.

**Key Features of Seaborn:**
- Built-in themes for improved plot design.
- Works well with DataFrames, making it easy to visualize datasets directly.
- Supports advanced visualizations such as violin plots, pair plots, and heatmaps.

### 2. **Plotting and Interpreting Various Types of Graphs**

Seaborn offers a wide variety of advanced plot types, from basic bar plots to complex heatmaps and multi-plot grids. Below are some of the key types of graphs, along with their interpretations.

#### a) **Scatter Plot**
A scatter plot shows the relationship between two numerical variables. It's useful for identifying trends, correlations, and outliers.


````python
# Scatter plot showing relationship between total bill and tip
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Total Bill vs Tip')
plt.show()
````

**Interpretation:** In this plot, we can see whether there is a linear relationship between the total bill and the tip amount. If the points trend upwards, it suggests a positive correlation.

#### b) **Line Plot**
A line plot is used to represent data points over a continuous range, such as time. Seaborn’s `lineplot` is especially useful for visualizing time series data.


````python
# Line plot
sns.lineplot(x='total_bill', y='tip', data=tips)
plt.title('Line Plot of Total Bill vs Tip')
plt.show()
````

#### c) **Bar Plot**
A bar plot is used to display the relationship between a categorical variable and a numerical variable. It’s excellent for comparing groups.


````python
# Bar plot of total bill based on different days
sns.barplot(x='day', y='total_bill', data=tips)
plt.title('Average Total Bill by Day')
plt.show()
````

**Interpretation:** The bar plot shows the average total bill on each day of the week, making it easy to compare how different categories (days) relate to the numerical variable (total bill).


#### d) **Box Plot**
A box plot (box-and-whisker plot) displays the distribution of a dataset based on five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. It helps in detecting outliers.

````python
# Box plot of total bill for different days
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Total Bill Distribution by Day')
plt.show()
````

**Interpretation:** Box plots are useful for visualizing the spread and skewness of the data. The middle line in the box represents the median, and the "whiskers" show the range of the data. Outliers are typically shown as individual points.

#### e) **Violin Plot**
A violin plot combines the benefits of box plots and kernel density plots. It shows the distribution of the data while also highlighting the probability density of the data at different values.

````python
# Violin plot of total bill by day
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title('Violin Plot of Total Bill by Day')
plt.show()
````

**Interpretation:** A violin plot allows us to understand both the distribution and the density of the data across different categories. Wider sections of the plot indicate higher probability density.

#### f) **Heatmap**
A heatmap is a graphical representation of data where individual values are represented by color. It’s particularly useful for showing correlations between variables in a dataset.

#### 1. Drop Non-Numeric Columns
If columns with non-numeric data aren't needed for the correlation matrix, simply drop them:

````python
# Heatmap of the correlation matrix
# Load example dataset from Seaborn
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

# Select only numeric columns
numeric_tips = tips.select_dtypes(include='number')
print(numeric_tips)
# Calculate correlation
corr = numeric_tips.corr()
print()

# Plot the heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

````

#### 2. Convert Categorical Data to Numeric (if relevant)
If categorical data (e.g., 'Yes'/'No' or 'Male'/'Female') is meaningful, encode it as numeric data:

````python
print(tips)
````

````python
# Convert categorical columns to numeric (if needed)
tips['sex'] = tips['sex'].map({'Male': 1, 'Female': 0})
tips['smoker'] = tips['smoker'].map({'Yes': 1, 'No': 0})
tips['day'] = tips['day'].astype('category').cat.codes
tips['time'] = tips['time'].map({'Lunch': 0, 'Dinner': 1})

# Now calculate the correlation
corr = tips.corr()

# Plot the heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
````

**Interpretation:** A heatmap visualizes correlations between numerical variables. The color intensity represents the strength of the correlation. The `annot=True` argument displays the correlation values on the plot.

#### g) **Pair Plot**
A pair plot visualizes the pairwise relationships and distributions of multiple variables. It plots scatter plots for numerical variables and histograms for their distributions.

````python
# Pair plot of the tips dataset
sns.pairplot(tips)
plt.show()
````

**Interpretation:** Pair plots help in visualizing relationships between multiple pairs of variables in one compact visualization. This is very helpful in detecting correlations and patterns across several variables.

#### h) **Facet Grid**
A Facet Grid allows you to plot multiple plots in a grid, with each plot representing a subset of the data based on some categorical variable(s). This is especially useful for visualizing complex relationships between variables.

````python
# Facet grid of scatter plots between total bill and tip, separated by sex
g = sns.FacetGrid(tips, col="sex")
g.map(sns.scatterplot, "total_bill", "tip")
plt.show()
````

**Interpretation:** This grid of scatter plots shows the relationship between total bill and tip for males and females separately, making it easy to spot differences between the two groups.

### Customization in Seaborn Plots

One of Seaborn's strengths is the ability to customize and fine-tune plots. Below are some common ways to customize your visualizations.

#### a) **Setting Plot Aesthetics**
Seaborn provides five different themes: darkgrid, whitegrid, dark, white, and ticks. You can set any of these themes to change the look of the plots.


````python
# Set the theme
sns.set_theme(style='whitegrid')
````

#### b) **Adding Titles and Labels**
While Seaborn automatically adds axes labels, adding custom titles and adjusting font sizes enhances the readability of plots.


````python
# Adding a title and labels to a plot
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Total Bill Distribution by Day', fontsize=16)
plt.xlabel('Day', fontsize=12)
plt.ylabel('Total Bill', fontsize=12)
plt.show()
````

#### c) **Adding Legends**
Seaborn allows you to create legends to clarify what the colors or markers represent.

````python
# Adding a legend to a plot
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=tips)
plt.legend(title='Gender')
plt.show()
````

### Using the Nigeria Economic data

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_excel('nigeria_econs_data.xlsx')
print(df)
````

### Lineplot showing the 4 economic variables

````python
# Line plot
import numpy as np
# Lineplot
sns.set(style="whitegrid", palette="deep")
plt.figure(figsize=(12, 8))

sns.lineplot(x = np.arange(1, 129), y = 'Nominal_GDP', data=df, label='Nominal GDP', linewidth=4, marker='o')
sns.lineplot(x = np.arange(1, 129), y = 'Real_GDP', data=df, label='Real GDP', linewidth=4, marker='s')
sns.lineplot(x = np.arange(1, 129), y = 'Imports', data=df, label='Imports', linewidth=3, marker='^')
sns.lineplot(x = np.arange(1, 129), y = 'Exports', data=df, label='Exports', linewidth=3, marker='D')

# Add titles and labels
plt.title("Trends of GDP, Imports, and Exports in Nigeria", fontsize=18, fontweight='bold')
plt.xlabel("Time", fontsize=14)
plt.ylabel("Values (in Million Naira)", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Customize legend
plt.legend(title="Variables", title_fontsize=14, fontsize=12, loc='upper left')

# Add gridlines for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Add padding to make the plot look better
plt.tight_layout()

# Display the plot
plt.show()
````

````python
# Lineplot
sns.set(style="whitegrid", palette="deep")
plt.figure(figsize=(14, 8))

# Create individual lines for each variable
sns.lineplot(x=np.arange(1, len(df) + 1), y='Nominal_GDP', data=df, label='Nominal GDP', linewidth=2.5, marker='o')
sns.lineplot(x=np.arange(1, len(df) + 1), y='Real_GDP', data=df, label='Real GDP', linewidth=2.5, marker='s')
sns.lineplot(x=np.arange(1, len(df) + 1), y='Imports', data=df, label='Imports', linewidth=2, marker='^')
sns.lineplot(x=np.arange(1, len(df) + 1), y='Exports', data=df, label='Exports', linewidth=2, marker='D')

# Add titles and labels
plt.title("Trends of GDP, Imports, and Exports Over Time", fontsize=18, fontweight='bold')
plt.xlabel("Time (Index)", fontsize=14)
plt.ylabel("Values (in Millions)", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Customize legend
plt.legend(title="Variables", title_fontsize=14, fontsize=12, loc='upper left')

# Add gridlines for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Add padding to make the plot look better
plt.tight_layout()

# Display the plot
plt.show()
````

````python
# Calculate the correlation for the 4 variables
corr = df.corr()

# Plot the heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
````

### Summary

**Seaborn** provides an intuitive and high-level API for creating attractive and informative visualizations. The library simplifies the creation of complex visualizations, enabling quick and insightful exploratory data analysis. Key plots include:

- **Scatter Plot**: Visualizes relationships between two continuous variables.
- **Line Plot**: Tracks changes over continuous variables such as time.
- **Bar Plot**: Compares categories against a numerical variable.
- **Box Plot**: Displays the distribution of numerical data and outliers.
- **Violin Plot**: Combines box plot and density plot to visualize distributions.
- **Heatmap**: Visualizes correlations between variables or a matrix of values.
- **Pair Plot**: Visualizes relationships between multiple pairs of variables.
- **Facet Grid**: Plots multiple subplots across different categories.

With these tools and techniques, you'll be able to produce sophisticated visual analyses that can help uncover trends, patterns, and outliers in your data.
