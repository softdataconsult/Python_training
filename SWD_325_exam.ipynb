```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Create DataFrame
data = {
    'Marketing_Spend': [200, 500, 300, 1000, 1500, 2500, 1000, 400, 3000, 700],
    'App_Downloads': [1500, 3000, 2000, 4500, 5000, 6000, 4000, 2500, 7500, 2800]
}
df = pd.DataFrame(data)

# Scatter plot
plt.scatter(df['Marketing_Spend'], df['App_Downloads'])
plt.title("App Downloads vs Marketing Spend")
plt.xlabel("Marketing Spend ($)")
plt.ylabel("App Downloads")
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Reshape X for sklearn
X = df[['Marketing_Spend']]
y = df['App_Downloads']

# Fit the regression model
model = LinearRegression()
model.fit(X, y)

# Get coefficients
intercept = model.intercept_
slope = model.coef_[0]

print("Intercept:", intercept)
print("Slope:", slope)
print("R-squared:", model.score(X, y))
```

```python
import scipy.stats as stats

# Data
usa = [1.4, 1.3, 1.5]
europe = [1.7, 1.8, 1.6]
asia = [2.0, 2.1, 2.2]
australia = [1.9, 1.8, 1.7]

# One-way ANOVA
f_stat, p_value = stats.f_oneway(usa, europe, asia, australia)

print("F-statistic:", f_stat)
print("p-value:", p_value)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Data
data = {
    'Articles': [5, 8, 6, 10, 7, 12, 9, 15, 6, 14],
    'Social': [10, 20, 15, 25, 12, 30, 22, 35, 18, 33],
    'Backlinks': [59, 75, 60, 100, 65, 120, 85, 150, 70, 130],
    'Traffic': [300, 450, 350, 500, 380, 600, 460, 700, 360, 650]
}

df = pd.DataFrame(data)

# Define features (X) and target (y)
X = df[['Articles', 'Social', 'Backlinks']]
y = df['Traffic']

# Fit model
model = LinearRegression()
model.fit(X, y)

# Extract coefficients
intercept = model.intercept_
coefficients = model.coef_
r2 = model.score(X, y)

print("Intercept:", intercept)
print("Coefficients:", coefficients)
print("R-squared:", r2)
```
